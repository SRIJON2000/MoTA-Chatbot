```python
import spacy
import numpy as np
import json
from sklearn.metrics.pairwise import cosine_similarity

# Download and load the pre-trained model
nlp = spacy.load("en_core_web_md")
```

```python
with open('/content/drive/MyDrive/RP Work/website.json') as file:
    data = json.load(file)
questions=[]
answers=[]
for intent in data['intents']:
    for pattern in intent['patterns']:
        questions.append(pattern)
        answers.append(np.random.choice(intent['responses']))
```

```python
def find_similar_question(user_input):
    # Vectorize the user input
    user_input_vec = nlp(user_input).vector
    # Compute the similarity between the user input and each question
    similarities = []
    for question in questions:
        question_vec = nlp(question).vector
        similarity = cosine_similarity([user_input_vec], [question_vec])[0][0]
        similarities.append(similarity)
    # Find the index of the most similar question
    most_similar_index = np.argmax(similarities)
    # Return the corresponding answer
    return answers[most_similar_index]

# Define a loop to continuously ask for user input and provide responses
while True:
    user_input = input("You: ")
    response = find_similar_question(user_input)
    print("Chatbot:", response)
```
